# Линейная регрессия

Из файла, где указаны цена и размер, нужно найти корреляцию между стоимостью и размером недвижимости

## 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

##

```python
data = pd.read_csv(r'C:\Users\Даша\Desktop\Data Sience Course\Python\Статистика в Python\real_estate_price_size.csv')
```

## 

### 

```python
print("Тип data:", type(data))
print("\nПервые 5 строк:")
print(data.head())
print("\nТипы данных в колонках:")
print(data.dtypes)
print("\nИнформация о данных:")
print(data.info())
```

```python
x1 = data['size']
y = data['price']
```

### Explore the data

```python
data
```

```python
data.describe()
```

### Regression itself

```python
x = sm.add_constant(x1)
results = sm.OLS(y, x).fit()
results.summary()
```

###

```python
plt.scatter(x1,y)
plt.xlabel('size', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.show
```
